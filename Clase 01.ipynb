## Atajos de teclado
Una vez seleccionada una celda

|Atajo|Acción|
|:---:|:-----|
|`B`|Crea celda **debajo**|
|`A`|Crea celda **arriba**|
|`D,D`|**Elimina** celda(s)|
|`Z`|**Regresa** una acción|
|`C`|**Copia** celda(s)|
|`X`|**Corta** celda(s)|
|`V`|**Pega** celda(s)|
|`M`|Convierte celda(s) a **Markdown**|
|`Y`|Convierte celda(s) a **Code**|
|`Ctrl + ENTER`|**Ejecuta** celda|
|`Alt + ENTER`|**Ejecuta** celda y crea una debajo|
|`1`|Convierte celda a **Markdown** con un título|
|`2`|Convierte celda a **Markdown** con un subtítulo|

**Notas**
- La extensión de los notebooks es `ipynb` por ser un *IPython* notebook
- Se pueden crear arhivos únicamente markdown, esos llevan la extensión `md` (`README`)
- Los notebooks se pueden exportar a diferentes formatos, entre ellos, a PDF.

# Otros comandos útiles

|Atajo|Acción|

|`W + D`| Regresa al escritorio|
|`W + L`| Bloquea computadora|
